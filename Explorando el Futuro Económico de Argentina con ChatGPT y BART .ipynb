### Empleamos scikit-learn para realizar la transformaci贸n polin贸mica y ajustar un modelo lineal a los datos existentes.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Datos
x = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
y = np.array([22.6, 19.8, 16.4, 14.7, 31.6, 27.2, 42.6, 24.7, 47.6, 53.8, 36.1, 50.9, 94.8, 211.4])

# Reshape de los datos para la regresi贸n polin贸mica
X = x.reshape(-1, 1)

# Transformaci贸n polin贸mica
grado_polinomio = 3
transformador_polinomio = PolynomialFeatures(degree=grado_polinomio)
X_polinomico = transformador_polinomio.fit_transform(X)

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X_polinomico, y)

# Predicci贸n para a帽os futuros
a帽os_futuros = np.array([2024, 2025]).reshape(-1, 1)
a帽os_futuros_polinomico = transformador_polinomio.transform(a帽os_futuros)
prediccion = modelo.predict(a帽os_futuros_polinomico)

# Gr谩fica de los datos y la predicci贸n
plt.scatter(x, y, label='Datos')
plt.plot(a帽os_futuros, prediccion, label='Predicci贸n', color='red')

# Anotar los valores de predicci贸n
for a帽o, pred in zip(a帽os_futuros.flatten(), prediccion):
    plt.annotate(f'{pred:.2f}%', (a帽o, pred), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Predicci贸n de Inflaci贸n en Argentina')
plt.xlabel('A帽o')
plt.ylabel('% de inflaci贸n')
plt.legend()
plt.show()

```
