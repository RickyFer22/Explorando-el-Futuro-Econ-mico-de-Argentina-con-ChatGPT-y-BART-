### Esta función se adapta de manera notable a los datos, como se evidencia claramente en el gráfico siguiente:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Datos
x = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
y = np.array([22.6, 19.8, 16.4, 14.7, 31.6, 27.2, 42.6, 24.7, 47.6, 53.8, 36.1, 50.9, 94.8, 211.4])

# Función polinómica de grado 3
def f(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

# Ajuste de la función a los datos
params, covariance = curve_fit(f, x, y)

# Generar valores predichos con la función ajustada
x_pred = np.linspace(min(x), 2025, 100)
y_pred = f(x_pred, *params)

# Configuración de seaborn
sns.set(style="whitegrid")

# Gráfica con seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, label='Datos')
sns.lineplot(x=x_pred, y=y_pred, label=f'Ecuación: {params[0]:.2f}x³ + {params[1]:.2f}x² + {params[2]:.2f}x + {params[3]:.2f}', color='red')
plt.title('Inflación %  anual desde 2010 en Argentina')
plt.xlabel('Año')
plt.ylabel('% de inflación')
plt.legend()
plt.show()

```

### *El sobreajuste de la ecuación en los datos provoca un aumento desmedido de la inflación, lo cual, en términos económicos, se daría una hiperinflación.*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Datos
x = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
y = np.array([22.6, 19.8, 16.4, 14.7, 31.6, 27.2, 42.6, 24.7, 47.6, 53.8, 36.1, 50.9, 94.8, 211.4])

# Función logística
def logistic_function(x, a, b, c, d):
    return a / (1 + np.exp(-b * (x - c))) + d

# Ajuste de la función a los datos
params, covariance = curve_fit(logistic_function, x, y, p0=[100, 0.1, 2015, 0], maxfev=10000)

# Generar valores predichos con la función ajustada
x_pred = np.linspace(min(x), 2025, 100)
y_pred = logistic_function(x_pred, *params)

# Configuración de seaborn
sns.set(style="whitegrid")

# Gráfica con seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, label='Datos')
sns.lineplot(x=x_pred, y=y_pred, label=f'Ecuación Logística: {params[0]:.2f} / (1 + e^(-{params[1]:.2f} * (x - {params[2]:.2f}))) + {params[3]:.2f}', color='red')
plt.title('Inflación %  anual desde 2010 en Argentina')
plt.xlabel('Año')
plt.ylabel('% de inflación')
plt.legend()
plt.show()

```

### Empleamos scikit-learn para realizar la transformación polinómica y ajustar un modelo lineal a los datos existentes.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Datos
x = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
y = np.array([22.6, 19.8, 16.4, 14.7, 31.6, 27.2, 42.6, 24.7, 47.6, 53.8, 36.1, 50.9, 94.8, 211.4])

# Reshape de los datos para la regresión polinómica
X = x.reshape(-1, 1)

# Transformación polinómica
grado_polinomio = 3
transformador_polinomio = PolynomialFeatures(degree=grado_polinomio)
X_polinomico = transformador_polinomio.fit_transform(X)

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X_polinomico, y)

# Predicción para años futuros
años_futuros = np.array([2024, 2025]).reshape(-1, 1)
años_futuros_polinomico = transformador_polinomio.transform(años_futuros)
prediccion = modelo.predict(años_futuros_polinomico)

# Gráfica de los datos y la predicción
plt.scatter(x, y, label='Datos')
plt.plot(años_futuros, prediccion, label='Predicción', color='red')

# Anotar los valores de predicción
for año, pred in zip(años_futuros.flatten(), prediccion):
    plt.annotate(f'{pred:.2f}%', (año, pred), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Predicción de Inflación en Argentina')
plt.xlabel('Año')
plt.ylabel('% de inflación')
plt.legend()
plt.show()

```


### En este código, se utiliza la regresión polinómica para predecir la inflación en Argentina en los años 2024 y 2025. 

### Pasos del Código:

1. **Datos de Entrada:** Se proporcionan los datos históricos de inflación en los años anteriores (2010-2023).

2. **Transformación Polinómica:** Los datos de entrada se transforman en un conjunto polinómico de grado 3, lo que permite modelar mejor la relación no lineal entre el año y el porcentaje de inflación.

3. **Entrenamiento del Modelo:** Se crea y entrena un modelo de regresión lineal con los datos transformados.

4. **Predicción:** Se utiliza el modelo entrenado para predecir la inflación en los años 2024 y 2025.

5. **Gráfica:** Los datos históricos se muestran en una gráfica junto con la predicción para visualizar la tendencia y las proyecciones futuras.

6. **Anotaciones:** Se añaden anotaciones a la gráfica para resaltar y etiquetar los valores predichos en 2024 y 2025.

Este enfoque proporciona una herramienta visual y cuantitativa para anticipar posibles escenarios económicos futuros en base a datos históricos.
